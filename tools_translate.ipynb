```python
from transformers import pipeline

translator = pipeline("translation", model="facebook/m2m100_418M")

sentences = ["i can go from feeling so hopeless to so damned hopeful just from being around someone who cares and is awake"]
translations = [translator(sentence, src_lang="en", tgt_lang="id") for sentence in sentences]
print(translations)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-en-id"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

sentences = ["i can go from feeling so hopeless to so damned hopeful just from being around someone who cares and is awake"]
inputs = tokenizer(sentences, return_tensors="pt", padding=True)
outputs = model.generate(**inputs)

translations = [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]
print(translations)
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

df_train = pd.read_csv('/content/val.txt', names=['Text', 'Emotion'], sep=';')

model_name = "Helsinki-NLP/opus-mt-en-id"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def translate_text(text, idx=None):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model.generate(**inputs)
    translation = tokenizer.decode(outputs[0], skip_special_tokens=True)

    if idx is not None:
        print(f"Processed row {idx + 1}: {text} -> {translation}")
    return translation

df_train['Translated_Text'] = [
    translate_text(row['Text'], idx) for idx, row in df_train.iterrows()
]

df_train.to_csv('/content/translated_val.txt', index=False, sep=';')

```
